```python
from tensorflow.keras.layers import Dense, Flatten, LSTM, Input
from tensorflow.keras.models import Sequential
from numpy import array
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import LSTM
from numpy import array
# define model

inputs1 = Input(shape=(5, 1))
lstm1, state_h, state_c = LSTM(units = 4, return_sequences = True, return_state=True)(inputs1)
model = Model(inputs=inputs1, outputs=[lstm1, state_h, state_c])

# define input data
data = array([0.1, 0.2, 0.3, 0.5, 0.6]).reshape((1,5,1))
# make and show prediction
print(model.summary())
print(model.predict(data))
```

```python
model = Sequential()
model.add(LSTM(units=5, return_sequences=False, input_shape=(7,1)))


# define input data
data_x = array([1, 2, 3, 4, 5, 6, 7]).reshape((1,7,1))
data_y = array([8])

model.compile(loss="mse", optimizer='adam') #Reminder: LOSS function is MSE but others can be used!
model.fit(data_x, data_y, epochs=5, verbose=1) 
# make and show prediction
print(model.summary())
print(model.predict(data_x))
```
